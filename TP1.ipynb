```python
import numpy.random as random
def tad(R,M):
    
    result = 0
    
    lst_tad = [0 for _ in range(M+1)] # on ajoute le compteur à la fin du TAD
    
    
    for i in range (0,R):
        lst_tad[-1] = lst_tad[-1] + 1 #on ajoute un au compteur
        acces = random.randint(M) # on teste un accès de manière aléatoire
        lst_tad[acces] = lst_tad[acces]+1 #on incrémente la valeur dans la cellule correspondante
    
    for j in range (len(lst_tad)-1): # on compte le nombre de case qui ont reçu une incrémentation
        if lst_tad[j] != 0:
            result = result +1 
    
    return print(result,lst_tad) 
```
